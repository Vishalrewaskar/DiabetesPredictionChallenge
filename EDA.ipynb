## Ordinal Encoding

```python
# smoking status
data['smoking_status'] = data['smoking_status'].map({
    'Never': 0,
    'Former': 1,
    'Current': 2
})

# education level
data['education_level'] = data['education_level'].map({
    'No formal': 0,
    'Highschool': 1,
    'Graduate': 2,
    'Postgraduate': 3
})

# income level
data['income_level'] = data['income_level'].map({
    'Low': 0,
    'Lower-Middle': 1,
    'Middle': 2,
    'Upper-Middle': 3,
    'High': 4
})
```

## One-Hot Encoding

```python
one_hot_cols = ['gender', 'ethnicity', 'employment_status']

data = pd.get_dummies(
    data,
    columns=one_hot_cols,
    drop_first=True
)
```
